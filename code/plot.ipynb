```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
policy= pd.read_csv('Het_final_easy/Easy Policy.csv')
value= pd.read_csv('Het_final_easy/Easy Value.csv')
q1= pd.read_csv('Hard Q-Learning L0.1 q0.0 E0.1.csv')
q2= pd.read_csv('Hard Q-Learning L0.1 q0.0 E0.3.csv')
q3= pd.read_csv('Hard Q-Learning L0.1 q0.0 E0.5.csv')
```

```python
q1.head()
```

```python
x1= value['iter']
x2= policy['iter']
y1= value['time']
y2= policy['time']

plt.plot(x1,y1, label= 'Value Iteration')
plt.plot(x2,y2, label='Policy Iteration')
plt.legend()
plt.xlabel('No. of iterations')
plt.ylabel('Time')
plt.title('Easy Problem- Iterations v/s Time')
plt.savefig('graph/easy_iter_time.png')
```

```python
x1= q1['iter']
x2= q2['iter']
y1= q1['convergence']
y2= q2['convergence']
x3= q3['iter']
y3= q3['reward']

# plt.plot(x1,y1, label= 'epsilon= 0.1')
# plt.plot(x2,y2, label='epsilon= 0.3')
plt.plot(x3,y3, label='epsilon= 0.5')
# plt.legend()
plt.xlabel('No. of iterations')
plt.ylabel('Reward')
plt.title('Hard Problem- Q-Learning Reward (alpha=0.1, epsilon=0.5)')
plt.savefig('graph/hard_q_iter_reward_custom.png')
```
